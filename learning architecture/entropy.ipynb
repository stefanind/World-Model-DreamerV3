Entropy: a measure of the amount of uncertainty (suprise) in a random variables outcome.  

- if you know what exactly will happen, entropy = 0
- if you don't know, entropy > 0
- formally, it measure the expected uncertainty of the outcome.
    - surprise of an event x is proportional to -logp(x)
    - rare events (low p(x)) more surprising (more entropy)
    - common events (high p(x)) less surprising
    - H(X) = E[-logp(X)] -> avg uncertainty in predicting X
    - E[-logp(X)] = integral p(x) * -logp(x) * dx (for continuous random variable)

Entropy quantifies how hard it is to predict a random variable BEFORE observing it.  

related to variance:
- variance measures how far values are from mean
- entropy measures how uncertain the entire distribution is, based on the probabilities, not the values.

variance = spread in space (values)
entropy  = spread in information (likelihoods)

variance looks only at second moments (squared deviations)
entropy considers ALL probabilities, i.e., how the pdf/pmf is arranged  

Because of normal distribution, var ~ entropy since the entropy is monotonically related to the variance from the formula.  
How they differ in a practical sense: both bimodal and uniform can have same variance, but the uniform is more uncertain because the outcome is evenly spread and unpredictable.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4, 4, 1000)
pdf = (1 / (1 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 0) / 1)**2) # normal dist
plt.plot(x, pdf)
```

```python
plt.plot(x, -np.log(pdf))
```

When x = 0, y = 1, meaning that the surprise is low (low output).  
When x = -4, y = 9, meaning that the surprise is high (high output).

```python
plt.plot(x, -np.log(pdf) * pdf)
```

the entropy weighed by the pdf: 
When x = 0, the entropy contribution is high, why?  
- p(0) is the most likely outcome (probability of 0.4)
- -logp(0) is the least surprising outcome (bell curve shows it is equal to 1)
- multiplying them together weighs the -logp(0) by 0.4 so it remains very high because of the weight of that value on the original distribution (p(0) = 0.4).
